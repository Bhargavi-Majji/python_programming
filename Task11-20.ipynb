```python
weight = 60
height = 1.65

bmi = weight / (height ** 2)

print("BMI =", round(bmi, 2))
```

```python
p = 10000
r = 5
t = 2

si = (p * r * t) / 100

print("Simple Interest =", si)
print("Total Amount =", p + si)
```

```python
p = 10000
r = 5
t = 2

amount = p * (1 + r/100) ** t
ci = amount - p

print("Compound Interest =", round(ci, 2))
print("Maturity Amount =", round(amount, 2))
```

```python
cp = 500
sp = 700

if sp > cp:
    print("Profit =", sp - cp)
elif cp > sp:
    print("Loss =", cp - sp)
else:
    print("No Profit No Loss")
```

```python
distance = 100
mileage = 20
fuel_price = 110

fuel_needed = distance / mileage
cost = fuel_needed * fuel_price

print("Fuel Cost =", cost)
```
